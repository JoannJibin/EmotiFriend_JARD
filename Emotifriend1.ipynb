```python
from gtts import gTTS
import os
from groq import Groq
import gradio as gr
from faster_whisper import WhisperModel

# Groq client
client = Groq(api_key="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")  # replace with your key

# Whisper for transcription
whisper_model = WhisperModel("base")

# Groq Chat Function
def generate_response(prompt):
    chat_completion = client.chat.completions.create(
        messages=[
            {"role": "system", "content": "You are a helpful assistant. Reply in a simple and expressive way if emojis are given."},
            {"role": "user", "content": prompt}
        ],
        model="llama3-70b-8192",
        temperature=0.7
    )
    return chat_completion.choices[0].message.content

# Text-to-speech
def speak_text(text, filename="output.mp3"):
    tts = gTTS(text=text)
    tts.save(filename)
    return filename

# Transcribe voice input
def transcribe_audio(audio_path):
    segments, _ = whisper_model.transcribe(audio_path)
    transcription = ""
    for segment in segments:
        transcription += segment.text
    return transcription.strip()

# Caption mode
def caption(emojis):
    return generate_response(f"Create an expressive Instagram caption for these emojis: {emojis}")

# Text to emoji
def text_to_emoji(text):
    return generate_response(f"Convert this text to emojis: {text}")

# Talk mode (emojis → expressive sentence)
def talk_from_emojis(emojis):
    reply = generate_response(f"Reply in an expressive way using these emojis, without any explanation: {emojis}")
    audio_file = speak_text(reply)
    return reply, audio_file

import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002700-\U000027BF"  # Dingbats
        u"\U000024C2-\U0001F251"  # Enclosed characters
        "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)


# Full voice input workflow
def voice_to_reply(audio_file):
    transcribed = transcribe_audio(audio_file)
    emojis = text_to_emoji(transcribed)
    reply = generate_response(f"Respond expressively to this emoji string: {emojis}")
    reply_no_emojis = remove_emojis(reply)
    audio = speak_text(reply_no_emojis)
    return transcribed, emojis, reply_no_emojis, audio


# Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("## 🎤 EmotiFriend – Emoji ↔ Text ↔ Voice")

    with gr.Tab("Instagram Caption"):
        emojis_input = gr.Textbox(label="Enter emojis:")
        caption_output = gr.Textbox(label="Generated Caption:")
        caption_button = gr.Button("Generate Caption")
        caption_button.click(caption, inputs=emojis_input, outputs=caption_output)

        text_input = gr.Textbox(label="Enter text to convert to emoji:")
        emoji_output = gr.Textbox(label="Emoji Result:")
        emoji_button = gr.Button("Convert to Emoji")
        emoji_button.click(text_to_emoji, inputs=text_input, outputs=emoji_output)

    with gr.Tab("Talk for Children"):
        emojis_input2 = gr.Textbox(label="Enter emojis:")
        talk_output = gr.Textbox(label="Generated Response:")
        talk_audio = gr.Audio(label="Voice Output", type="filepath")
        talk_button = gr.Button("Talk")
        talk_button.click(talk_from_emojis, inputs=emojis_input2, outputs=[talk_output, talk_audio])

    with gr.Tab("Voice Input"):
      voice_input = gr.Audio(type="filepath", label="Upload your voice")  # Fixed
      transcribed_text = gr.Textbox(label="Transcribed Text")
      voice_emojis = gr.Textbox(label="Converted Emojis")
      voice_reply = gr.Textbox(label="AI Reply")
      voice_audio = gr.Audio(label="Voice Output", type="filepath")
      process_button = gr.Button("Process Voice")
      process_button.click(voice_to_reply, inputs=voice_input,
                         outputs=[transcribed_text, voice_emojis, voice_reply, voice_audio])


demo.launch()
```
